```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1.line

```python
data = pd.Series(np.random.randn(10).cumsum(),index = np.arange(0,100,10))
data.plot(xticks=data.index)
```

```python
df = pd.DataFrame(np.random.randn(10,4),columns=['A','B','C','D'],index=np.arange(0,100,10))
df.plot()
```

# 2.Bar & Barh

```python
fig,axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(16),index=list('abcdefghighlmnop'))
data.plot.bar(ax=axes[0],color='k',alpha=0.5)
data.plot.barh(ax=axes[1],color='k',alpha=0.5)
```

```python
df = pd.DataFrame(np.random.rand(6,4),columns=['A','B','C','D'],index=['one','two','three','four','five','six'])
df.plot.bar()
```

```python
df.T.plot.bar()
```

```python
_, axes = plt.subplots(1,2,figsize=(16,6))
df.plot.bar(ax=axes[0])
df.plot.bar(ax=axes[1],stacked=True)
```
